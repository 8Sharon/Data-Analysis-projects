```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#load the csv dataset
netflix_df=pd.read_csv('C:/Users/SHARON/Desktop/titanic/netflix_titles.csv')
netflix_df.head(11)
```

```python
netflix_df.info()
```

```python
netflix_df.describe()
```

```python
sns.countplot(netflix_df['type'])
fig = plt.gcf()
fig.set_size_inches(5,5)
plt.title('Type')
```

```python
#rating of shows and movies
sns.countplot(netflix_df['rating'])
sns.countplot(netflix_df['rating']).set_xticklabels(sns.countplot(netflix_df['rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(6,6)
plt.title('Rating')
```

```python
#relation between type and rating
plt.figure(figsize=(6,4))
sns.countplot(x='rating',hue='type',data=netflix_df)
plt.title('Relation between Type and Rating')
plt.show()
```

```python
#movie and tv shows
labels = ['Movie', 'TV show']
size = netflix_df['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (3, 3)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

```python
#Rating piechart
netflix_df['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(6,4))
plt.show()
```
