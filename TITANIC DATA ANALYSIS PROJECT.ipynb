```python
#load csv file into a data frame
titanic_df=pd.read_csv('C:/Users/SHARON/Downloads/titanic.csv')
```

```python
Survivors=titanic_df.groupby('Survived')['Age']
titanic_df.head(11)
```

```python
ax1, ax2=Survivors.plot(kind='hist', legend=True, alpha=0.6, bins=range(0, 90, 10))
ax1.set_xlabel('Age (Years)')
ax1.legend(['No', 'Yes'], title='Survived', loc='upper right')
ax2.set_ylabel('Frequency')
plt.title('Comparison of survivors v/s non survivors based on Age')
```

```python
#survival based on gender
Survivors=titanic_df.groupby('Survived')['Sex']
Survivors.describe()
```

```python
Survivors=pd.crosstab(titanic_df['Survived'], titanic_df['Sex'])
Survivors.plot(kind='bar', legend=True)
plt.xlabel('Survived')
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.ylabel('Frequency')
plt.legend(['Female', 'Male'], title='Sex', loc='upper right')
plt.title('Comparison of survivors v/s non survivors based on sex')
```

```python
#survival based on ticket class
Survivors=titanic_df.groupby('Survived')['Pclass']
Survivors.describe()
```

```python
Survivors=pd.crosstab(titanic_df['Survived'], titanic_df['Pclass'])
Survivors.plot(kind='bar')
plt.xlabel('Survived')
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.ylabel('Frequency')
plt.legend(['1st', '2nd', '3rd'], title='Ticket Pclass', loc='upper right')
plt.title('Comparison of survivors v/s non survivors based on Ticket Pclass')
```

```python
#survival based on siblings and spouses aboard
Survivors=titanic_df.groupby('Survived')['SibSp']
Survivors.describe()
```

```python
ax1, ax2=Survivors.plot(kind='hist', legend=True, alpha=0.6, bins=range(0, 8, 1))
ax1.set_xlabel('No. of siblings and spouses Aboard')
ax1.legend(['No', 'Yes'], title='Survived', loc='upper right')
ax2.set_ylabel('Frequency')
plt.title('Comparison of survivors v/s non survivors based on based on siblings and spouses')
```
